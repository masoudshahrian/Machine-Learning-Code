```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python

# تعریف داده‌های آموزشی
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
# ساخت مدل
model = Sequential()
model.add(Dense(5, activation='relu', input_shape=(2,)))  # لایه پنهان با 2 واحد و تابع فعالسازی relu
model.add(Dense(50, activation='relu'))  # لایه پنهان با 2 واحد و تابع فعالسازی relu
model.add(Dense(30, activation='relu'))  # لایه پنهان با 2 واحد و تابع فعالسازی relu
model.add(Dense(1, activation='sigmoid'))  # لایه خروجی با 1 واحد و تابع فعالسازی sigmoid
# کامپایل مدل
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.1), metrics=['accuracy'])
# آموزش مدل
model.fit(X, y, epochs=10, verbose=0)  # آموزش مدل به مدت 100 دوره
# ارزیابی مدل
score = model.evaluate(X, y, verbose=0)
print('دقت:', score[1])

```

```python
#Neural Network
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

# Load the iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create a neural network classifier
nn = MLPClassifier()

# Fit the classifier to the training data
nn.fit(X_train, y_train)

# Evaluate the classifier on the test data
accuracy = nn.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Predict the class labels of the test data
y_pred = nn.predict(X_test)

# Print the predicted class labels
print(f"Predicted class labels: {y_pred}")
print(f"True      class labels: {y_test}")
```

```python
x_new=[5,2,5,.1]
# Predict the class label of the new data
y_pred = nn.predict([x_new])

# Print the predicted class labels
print(f"Predicted class labels: {y_pred}")
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense

# Load the iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create a neural network model with Dense layers
model = Sequential()
model.add(Dense(5, activation='relu', input_shape=(4,)))  # First hidden layer with 4 neurons
model.add(Dense(100, activation='relu'))  # Second hidden layer with 4 neurons
model.add(Dense(30, activation='relu'))  # Third hidden layer with 4 neurons
model.add(Dense(20, activation='relu'))  # Fourth hidden layer with 4 neurons
model.add(Dense(3, activation='softmax'))  # Output layer with 3 units for classification
#model.add(Dense(3, activation='sigmoid'))  # Output layer with 3 units for classification

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model to the training data
model.fit(X_train, y_train, epochs=300, batch_size=32, verbose=0)

# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, y_test)
print("Accuracy:", accuracy)
```

با استفاده از لایه‌های `Dense` از کتابخانه Keras، میتوانید شبکه عصبی خود را با لایه‌های پرسپترون چندلایه دوباره بنویسید. الگوریتم زیر نشان می‌دهد که چگونه این کار انجام می‌دهیم.
در این الگوریتم، ما از مدل `Sequential` استفاده کرده و به ترتیب لایه‌های Dense را اضافه می‌کنیم. هر لایه با استفاده از تابع فعال‌سازی ReLU متصل می‌شود. در لایه خروجی، از تابع فعال‌سازی softmax برای دسته‌بندی استفاده می‌شود، زیرا ما با داده‌های چند دسته‌ای کار می‌کنیم.

در نهایت، مدل را با تابع هدایت کننده Adam و تابع هزینه 'sparse_categorical_crossentropy' کامپایل کرده و بر روی داده‌های آموزشی مدل را آموزش می‌دهیم. سپس دقت مدل را بر روی داده‌های تست محاسبه و چاپ می‌کنیم.

```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# Load the iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Create a k-NN classifier with 5 neighbors
knn = KNeighborsClassifier(n_neighbors=3)

# Fit the classifier to the training data
knn.fit(X_train, y_train)

# Evaluate the classifier on the test data
accuracy = knn.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Predict the class labels of the test data
y_pred = knn.predict(X_test)

# Print the predicted class labels
print(f"Predicted class labels: {y_pred}")
print(f"True      class labels: {y_test}")
```

# Making Predictions
We can now make predictions using this model on new data for which we might not know the correct labels. Imagine we found an iris in the wild with a sepal length of 5 cm, a sepal width of 2.9 cm, a petal length of 1 cm, and a petal width of 0.2 cm.
What species of iris would this be? We can put this data into a NumPy array, again by calculating the shape—that is, the number of samples (1) multiplied by the number of features (4):

```python
import numpy as np
X_new = np.array([[4, 1, 1, 0.02]])
print("X_new.shape: {}".format(X_new.shape))

#To make a prediction, we call the predict method of the knn object:
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(iris['target_names'][prediction]))
```

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# load the playtennis dataset
data = pd.read_csv('PlayTennisData.csv')

# encode the categorical data
le = LabelEncoder()
data = data.apply(le.fit_transform)

# split the data into features and target
x = data.drop('PlayTennis', axis=1)
y = data['PlayTennis']

# split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# create a gaussian naive bayes classifier
gnb = GaussianNB()

# fit the classifier to the training data
gnb.fit(x_train, y_train)

# evaluate the classifier on the test data
accuracy = gnb.score(x_test, y_test)
print(f"accuracy: {accuracy:.2f}")
```

```python
In addition to Gaussian Naive Bayes, there are several other types of Naive Bayes classifiers, including:
- **Multinomial Naive Bayes**: This type of Naive Bayes classifier is commonly used for text classification, where the data is typically represented as word counts or term frequencies².
- **Bernoulli Naive Bayes**: This type of Naive Bayes classifier is similar to the Multinomial Naive Bayes classifier, but it is designed for binary/boolean features².
- **Complement Naive Bayes**: This type of Naive Bayes classifier is an adaptation of the standard Multinomial Naive Bayes classifier and is particularly suited for imbalanced data sets².
```

```python
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load the iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create a Random Forest classifier
rf = RandomForestClassifier()

# Fit the classifier to the training data
rf.fit(X_train, y_train)

# Evaluate the classifier on the test data
accuracy = rf.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")
```
