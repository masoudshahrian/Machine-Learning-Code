```python
import numpy as np
import sklearn.linear_model as li
# import sklearn.metrics as met
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
plt.style.use('ggplot')

nD=1000
nX=3
X=np.zeros((nD, nX))
Y=np.zeros((nD, 1))

for i in range (nD):
     X[i,:]=np.random.uniform(-1,1,nX)
     Y[i,0]=1.2*X[i,0]-0.9*X[i,1]+0.4*X[i,2]-0.5
```

```python
X
```

```python
Y
```

```python
Model = li.LinearRegression()
Model.fit(X, Y)
O= Model.predict(X)
```

```python
O
```

```python

#  Show Result
plt.scatter(Y[:,0], O[ : , 0], s=12, color='blue',label='RealData')
plt.plot([-3,2],[-3,2],lw=1.2, color='red', label='X=Y')
plt.title('Linear Regression')
plt.xlabel('TargetValue')
plt.ylabel('PredictedValue')
plt.legend()
plt.show()




```
